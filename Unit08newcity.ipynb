```python
import matplotlib as mpl
import matplotlib.pyplot as plt
 
%config InlineBackend.figure_format = 'retina'
 
!apt -qq -y install fonts-nanum
 
import matplotlib.font_manager as fm

fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath, size=9)
plt.rc('font', family='NanumBarunGothic') 
mpl.font_manager._rebuild()
```

```python
import csv
f = open('inpast.csv', encoding ='cp949')
data = csv.reader(f)
mp = []
fp = []

for row in data : 
  if '연희동' in row[0]:
    for i in row[3:104]:    #남성 데이터 : 3~103번 데이터 슬라이싱 - 정수 데이터 변환 - m 리스트에 저장
      mp.append(-int(i.replace(',','')))
    for i in row[106:]:     #여성 데이터 : 106번~끝까지 데이터 슬라이싱 - 정수 데이터 변환 - f 리스트에 저장
      fp.append(int(i.replace(',','')))

import matplotlib.pyplot as plt

plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background
plt.rcParams['axes.unicode_minus'] = False
plt.title('2011년 인천 연희동의 남녀 성별 인구 분포')
plt.barh(range(101), mp,color = 'r', label = '남성') # Make a horizontal bar plot
plt.barh(range(101), fp , color ='b', label= '여성') # Make a horizontal bar plot
plt.legend()
plt.show()
```

```python
import csv
f = open('innow.csv', encoding ='cp949')
data = csv.reader(f)
mn = []
fn = []

for row in data : 
  if '연희동' in row[0]:
    for i in row[3:104]:    #남성 데이터 : 3~103번 데이터 슬라이싱 - 정수 데이터 변환 - m 리스트에 저장
      mn.append(-int(i.replace(',','')))
    for i in row[106:]:     #여성 데이터 : 106번~끝까지 데이터 슬라이싱 - 정수 데이터 변환 - f 리스트에 저장
      fn.append(int(i.replace(',','')))
import matplotlib.pyplot as plt

plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background
plt.rcParams['axes.unicode_minus'] = False
plt.title('2021년 인천 연희동 남녀 성별 인구 분포')
plt.barh(range(101), mn, color = 'r', label = '남성') # Make a horizontal bar plot
plt.barh(range(101), fn, color ='b', label= '여성') # Make a horizontal bar plot
plt.legend()
plt.show()
```

```python
plt.figure(dpi=100)
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background
plt.rc('font', family='NanumBarunGothic') 
plt.title("2011, 2021 인천신도시 남녀 인구 분포")
plt.barh(range(101),mn,color='r',label='2021_남')
plt.barh(range(101),fn,color='b',label='2021_여')
plt.barh(range(101),mp,color='pink',label='2011_남')
plt.barh(range(101),fp,color='skyblue',label='2011_여')
plt.xlabel('인구 수')
plt.ylabel('나이')
plt.legend(loc='upper left')
plt.show()
```
