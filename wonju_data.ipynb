```python
import csv
import matplotlib.pyplot as plt

f=open('Wonju.csv',encoding='cp949')
data=csv.reader(f)
next(data)

max_temp=[]
min_temp=[]
mean_temp=[]

for row in data:
  if row[-1] != '':
    max_temp.append(float(row[-1]))
  if row[-2] !='':
    min_temp.append(float(row[-2]))
  if row[-3] != '':
    mean_temp.append(float(row[-3]))

plt.figure(dpi=200)
plt.hist(max_temp,bins=30,color='r',label='max')
plt.hist(min_temp,bins=30,color='b', label = 'min')
plt.hist(mean_temp,bins=30,color='g', label='mean')
plt.title('Temperature in Wonju while YoungWook in Military')
plt.legend()
plt.show()
    
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data)

max_temp_Wonju=[]

for row in data:
  if row[-1] != '':
    max_temp_Wonju.append(float(row[-1]))

f=open('Wonju.csv',encoding='cp949')
data=csv.reader(f)
next(data)

max_temp_Seoul=[]

for row in data:
  if row[-1] != '':
    max_temp_Seoul.append(float(row[-1]))


plt.figure(dpi=200)
plt.boxplot([max_temp_Wonju, max_temp_Seoul])
plt.xticks([1,2],['wonju', 'seoul'])
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('Wonju.csv',encoding='cp949')
data=csv.reader(f)
next(data)

month=[]
for i in range(12):
  month.append([])

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=200)
plt.boxplot(month, showfliers=False)
plt.show()
```

```python
#   
import csv
import matplotlib.pyplot as plt

f=open('Wonju.csv',encoding='cp949')
data=csv.reader(f)
next(data)

day =[]
for i in range(31):
  day.append([])

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '06':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.figure(figsize=(10,5),dpi=200)
plt.style.use('ggplot')
plt.boxplot(day, showfliers=False)
plt.show()
```
