자전거 대여소 공공데이터 준비하기

```python
# 1. pandas 선언
import pandas as pd

#2. CSV 파일 불러오기\
df = pd.read_csv('세종특별자치시_공영자전거 어울링 대여소_20200915.csv', encoding='cp949')
df.head()
```

지도에 표시할 대여소 정보 추출

```python
df_route = df.loc[:,['STATION_NM', 'AREA', 'Y_POS', 'X_POS']]
df_route.head()
```

자전거 대여서 Marker 활용하여 지도 표출

```python
# 지도에 표시하기 위해 folium 라이브러리
import folium

map = folium.Map(location=[df_route['Y_POS'].mean(), df_route['X_POS'].mean()],
                 zoom_start=11)
for n in df_route.index:
  folium.Marker([df_route.loc[n, 'Y_POS'], df_route.loc[n,'X_POS']],    # 위도, 경도
                popup=df_route.loc[n,'STATION_NM'] + ' ' +df_route.loc[n, 'AREA'],
                icon=folium.Icon(color='green', icon = 'fa-bicycle', prefix='fa')).add_to(map)
    
map
```

가로등 설치 공공데이터 시각화

```python
map = folium.Map(location=[df_route['Y_POS'].mean(), df_route['X_POS'].mean()],
                 zoom_start=11)

for n in df_route.index:

  if df_route.loc[n, 'AREA'] == '1생활권' :
    folium.Marker([df_route.loc[n,'Y_POS'], df_route.loc[n, 'X_POS']], #위도, 경도
                  popup=df_route.loc[n, 'STATION_NM']+' '+df_route.loc[n, 'AREA'],
                  icon=folium.Icon(color='green', icon='fa-bicycle', prefix='fa')).add_to(map)

  elif df_route.loc[n, 'AREA'] == '2생활권' :
    folium.Marker([df_route.loc[n,'Y_POS'], df_route.loc[n, 'X_POS']], #위도, 경도
                  popup=df_route.loc[n, 'STATION_NM'] + ' ' +df_route.loc[n, 'AREA'],
                  icon=folium.Icon(color='red', icon='fa-bicycle', prefix='fa')).add_to(map)

  elif df_route.loc[n, 'AREA'] == '3생활권' :
    folium.Marker([df_route.loc[n,'Y_POS'], df_route.loc[n, 'X_POS']], #위도, 경도
                  popup=df_route.loc[n, 'STATION_NM'] + ' ' +df_route.loc[n, 'AREA'],
                  icon=folium.Icon(color='black', icon='fa-bicycle', prefix='fa')).add_to(map)

  elif df_route.loc[n, 'AREA'] == '4생활권' :
    folium.Marker([df_route.loc[n,'Y_POS'], df_route.loc[n, 'X_POS']], #위도, 경도
                  popup=df_route.loc[n, 'STATION_NM'] + ' ' +df_route.loc[n, 'AREA'],
                  icon=folium.Icon(color='#c32f83', icon='fa-bicycle', prefix='fa')).add_to(map)

  else:
    folium.Marker([df_route.loc[n,'Y_POS'], df_route.loc[n, 'X_POS']], #위도, 경도
                  popup=df_route.loc[n, 'STATION_NM'] + ' ' +df_route.loc[n, 'AREA'],
                  icon=folium.Icon(color='#ff6939', icon='fa-bicycle', prefix='fa')).add_to(map)
```

가로등 설치 공공데이터 준비하기
