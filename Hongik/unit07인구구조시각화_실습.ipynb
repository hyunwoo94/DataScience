```python
import csv

f = open('age.csv', encoding='cp949')
data = csv.reader(f)

result=[]

for row in data:
  if '역삼2동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',',''))) #인구수 데이터 정수로 변환 후 result 변수에 담기 #1,000단위 , 제거후 int형
print(result)
```
