```python
import matplotlib.pyplot as plt
import csv
f= open('gender.csv',encoding='cp949')
data =csv.reader(f)
result_busan=[] #남여 차이 비율 퍼센트 담을 리스트
name_busan='부산광역시'
for row in data:
  if name_busan in row[0]:
    for i in range(3,104) :
       if i%10 == 0:
         result_busan.append((int(row[i].replace(',',''))-int(row[i+103].replace(',','')))/(int(row[i].replace(',',''))+int(row[i+103].replace(',',''))))
    break
result_busan=[i*100 for i in result_busan]
f.close()
print(len(result_busan))
```
