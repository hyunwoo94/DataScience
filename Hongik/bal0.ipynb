```python
import csv
import matplotlib.pyplot as plt


f = open('Wonju.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터를 저장할 리스트 생성
result1 = []  # 최저
result2 = []   # 평균


for row in data:
  if row[-1] !='':
    result.append(float(row[-1]))
  if row[-2] !='':
    result1.append(float(row[-2]))
  if row[-1] !='':
    result2.append(float(row[-3]))

plt.figure(dpi = 200)
plt.boxplot([result,result1,result2])


plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('Wonju.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) #헤더

high_temp = [] #  최고 기온을 담은 리스트 변수 선언
low_temp = [] #  최저 기온을 담을 리스트 변수 선언
mean_temp = []

for row in data:
  month = row[0].split('-')[1]

  if row[-1]  != '':
    high_temp.append(float(row[-1])) 
  if row[-2]  != '':
    low_temp.append(float(row[-2]))
  if row[-3]  != '': 
    mean_temp.append(float(row[-3]))

plt.figure(dpi=200)
plt.hist(high_temp, bins=30, color='r', label="High_temp")
plt.hist(low_temp, bins=30, color='b',label ="Low_temp")
plt.hist(mean_temp, bins=30, color='g',label ="Mean_temp")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('Wonju.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

month = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(12) :
  month.append([]) # month 리스트 내에 18개 리스트 생성 [[],[],[],[],[],[],............31개]

for row in data:
   if row[-1] != "":
       month[int(row[0].split('-')[1])-1].append(float(row[-1]))


plt.figure(dpi = 200)
plt.boxplot(month, showfliers= False)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('Wonju.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31) :
  day.append([]) # day 리스트 내에 31개 리스트 생성 [[],[],[],[],[],[],............31개]

for row in data:
   if row[-1] != "":
     if row[0].split('-')[1] == '08':
       day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot') # 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(figsize=(10,5),dpi = 200)
plt.boxplot(day, showfliers =False) # 아웃라이어(이상치) 값 생략
plt.show()
```
