인구 공공데이터 준비하기

```python
import csv

f= open('age.csv', encoding = 'cp949')
data = csv.reader(f)

result=[] #우리동네 연령별 데이터 결과

for row in data:
  if '서울특별시 강남구 역삼2동' in row[0]:
    for i in row[3:]:
      result.append(i)
print(result)
```

우리 동네 연령별 인구수 result 리스트에 저장하기 -> 시각화를 위해

```python
import csv

f = open('age.csv', encoding='cp949')
data = csv.reader(f)

result=[]

for row in data:
  if '신도림' in row[0]:
    for i in row[3:]:
      result.append(int(i)) #인구수 데이터 정수로 변환 후 result 변수에 담기
print(result)
```

```python
import matplotlib.pyplot as plt

plt. figure(dpi = 200)
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background

plt.plot(result)
plt.show()
```
