1.데이터 읽어오기

```python
import pandas as pd

df = pd.read_csv('age.csv', index_col=0)
df.head()
```

```python
df.info()
```

```python
df.head()
```

인구수 -> 인구비율로 전처리

```python
df = df.div(df['총인구수'], axis=0)  # 전체 데이터를 총 인구수로 나눠서 비율로 변환
df.head()
```

```python
del df['총인구수'], df['연령구간인구수']  # 총인구수, 연령구간인구수 열삭제
df.head()
```

2~3. 궁금한 지역 이름 받고 해당 지역의 인구 구조 저장하기

```python
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ') # 지역이름 입력
a = df.index.str.contains(name) # 해당 행의 인덱스를 찾기

df2 = df[a] #인구 구조를 저장
df2.head()
```

```python
df2.T
```

궁금한 지역 인구구조 시각화

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10,4)
plt.rcParams['figure.dpi'] = 200

plt.style.use('ggplot')
df2.T.plot()
plt.show()
```

```python
df.sub(df2.iloc[0],axis = 1)  # home - away
```

```python
df2.iloc[0]
```

```python
df.sub(df2.iloc[0], axis=1).sum(axis=1).sort_values().index[:5]
```

```python
import numpy as np 

df.loc[np.power(df.sub(df2.iloc[0], axis=1),2).sum(axis=1).sort_values().index[:5]]
```

```python
import numpy as np

df.loc[np.power(df.sub(df2.iloc[0], axis=1),2).sum(axis=1).sort_values().index[:5]].T.plot()
plt.show()
```
