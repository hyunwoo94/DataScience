```python
#최고 기온 데이터만 리스트에 담아서 출력하기
import csv
f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] !='':
    result.append(float(row[-1]))
  
print(result) #최고 기온 리스트 출력
```

```python
# 최고기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화
#최고 기온 데이터만 리스트에 담아서 출력하기
import csv
f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] != '':  # 최고 기온 결측치 제외
    if row[0].split('-')[1] =='08':  #날짜를 잘라서 8월 데이터 추출
      result.append(float(row[-1])) #결과 리스트에 담기

plt.figure(dpi=200)
plt.plot(result, 'lightpink')  # result 리스트의 최고 기온 데이터를 빨간색 그래프로 그리기
plt.show()
```

매년 생일 기준을 최고 기온 그래프 그리기

```python
# 최고기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화
#최고 기온 데이터만 리스트에 담아서 출력하기
import csv
f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] != '':  # 최고 기온 결측치 제외
    if row[0].split('-')[1] =='02' and row[0].split('-')[2] =='22': #날짜를 잘라서 8월 데이터 추출
      result.append(float(row[-1])) #결과 리스트에 담기

plt.figure(dpi=200)
plt.plot(result, 'lightpink')  # result 리스트의 최고 기온 데이터를 빨간색 그래프로 그리기
plt.show()
```

매년 생일 기준 최고기온, 최저기온 그래프 그리기



```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) #헤더

high =[]  # 최고 기온
low = []  # 최저 기온

for row in data:
  if row[-1] !='' and row[-2] !='':
    if row[0].split('-')[1] == '02' and row[0].split('-')[2] =='22':
      high.append(float(row[-1]))  # 최고기온
      low.append(float(row[-2]))  # 최저기온

plt.figure(dpi=200)
plt.plot(high, 'g', label ='high')  # 최고기온
plt.plot(low, 'r', label ='low')  # 최저기온
plt.legend()  #범례 표시
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) #헤더

high = []
low =[]

for row in data:
  if row[-1] !='' and row[-2] !='':
    if 1983 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1] =='06' and row[0].split('-')[2] =='09':
        high.append(float(row[-1]))  # 최고기온
        low.append(float(row[-2]))  # 최저기온

#데이터 시각화
plt.figure(dpi=200)
plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')

plt.legend() # 범례 표시
plt.show()
```

매년 생일 평균기온

```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) #헤더

high = []
low =[]
avg =[] #평균기온

for row in data:
  if row[-1] !='' and row[-2] !='' and row[-3] !='': # 최고, 최저, 평균 결측치 제거
    if 1983 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1] =='06' and row[0].split('-')[2] =='09':
        high.append(float(row[-1]))  # 최고기온
        avg.append(float(row[-3]))
        low.append(float(row[-2]))  # 최저기온

#데이터 시각화
plt.figure(dpi=200)
plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')
plt.plot(avg, 'g', label= '평균기온')
plt.title('내 생일의 기온 변화 그래프')
plt.legend() # 범례 표시
plt.show()
```
